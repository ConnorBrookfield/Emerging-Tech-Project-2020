```python
import pandas as pd
import numpy as np
import tensorflow
from tensorflow import keras as kr
from tensorflow.keras import layers

# Reading in the csv file and assigning column labels
windturbine_csv = pd.read_csv(
    "powerproduction.csv",
    names=["speed", "power"])

train_data = windturbine_csv.copy()

# Splitting the Data into Training Collumns
power_data_train = train_data.pop("power")
speed_data_train = train_data.pop("speed")

# Model Creation
model = kr.models.Sequential()

# Adding a Flatten layer, and 3 dense layers to the model
model.add(layers.Flatten())
model.add(layers.Dense(50, input_shape=(1,), activation='sigmoid', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
model.add(layers.Dense(100, input_shape=(1,), activation='sigmoid', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
model.add(layers.Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))

# Compiling the model using an "Adam" optimizer and a "Mean squared error" loss algorithm 
model.compile(kr.optimizers.Adam(lr=0.001), loss='mean_squared_error')
model.fit(speed_data_train, power_data_train, epochs=100, batch_size=10)
```

```python
# Setting a wind speed that we want to predict the power output for
wind_speed = [0]
print(wind_speed)

# Testing the accuracy of the predictions (using the "wind_speed" guess)
model.predict(wind_speed)

```
